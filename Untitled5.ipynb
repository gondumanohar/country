```python
data = pd.read_csv(r'C:\Users\manohar\OneDrive\Desktop\country_comparison_large_dataset.csv')

X = data[['Year','Population (in Millions)','Population Growth Rate (%)']] 
y = data['Unemployment Rate (%)'] 


threshold = y.median()  
y_binary = (y >= threshold).astype(int)  


X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.2, random_state=42)

logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
y_pred_logistic = logistic_model.predict(X_test)
y_pred_logistic_proba = logistic_model.predict_proba(X_test)[:, 1] 


decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(X_train, y_train)
y_pred_decision_tree = decision_tree_model.predict(X_test)
y_pred_decision_tree_proba = decision_tree_model.predict_proba(X_test)[:, 1]

random_forest_model = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest_model.fit(X_train, y_train)
y_pred_random_forest = random_forest_model.predict(X_test)
y_pred_random_forest_proba = random_forest_model.predict_proba(X_test)[:, 1]


print("Logistic Regression:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_logistic):.2f}")
print("Classification Report:\n", classification_report(y_test, y_pred_logistic))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_logistic))
print(f"ROC-AUC: {roc_auc_score(y_test, y_pred_logistic_proba):.2f}")


print("\nDecision Tree Classifier:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_decision_tree):.2f}")
print("Classification Report:\n", classification_report(y_test, y_pred_decision_tree))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_decision_tree))
print(f"ROC-AUC: {roc_auc_score(y_test, y_pred_decision_tree_proba):.2f}")

print("\nRandom Forest Classifier:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_random_forest):.2f}")
print("Classification Report:\n", classification_report(y_test, y_pred_random_forest))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_random_forest))
print(f"ROC-AUC: {roc_auc_score(y_test, y_pred_random_forest_proba):.2f}")


plt.figure(figsize=(10, 8))


fpr_logistic, tpr_logistic, _ = roc_curve(y_test, y_pred_logistic_proba)
plt.plot(fpr_logistic, tpr_logistic, label="Logistic Regression (AUC = {:.2f})".format(roc_auc_score(y_test, y_pred_logistic_proba)))


fpr_dt, tpr_dt, _ = roc_curve(y_test, y_pred_decision_tree_proba)
plt.plot(fpr_dt, tpr_dt, label="Decision Tree (AUC = {:.2f})".format(roc_auc_score(y_test, y_pred_decision_tree_proba)))


fpr_rf, tpr_rf, _ = roc_curve(y_test, y_pred_random_forest_proba)
plt.plot(fpr_rf, tpr_rf, label="Random Forest (AUC = {:.2f})".format(roc_auc_score(y_test, y_pred_random_forest_proba)))

plt.plot([0, 1], [0, 1], 'k--') 
plt.title('ROC Curve Comparison')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='best')
plt.show()
```
